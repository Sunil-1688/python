#### Chart - 7 - Months Data

````python

# Chart - 7 visualization code
event_labels = {0: 'call', 1: 'sms', 2: 'data'}
df['item_label'] = df['item'].map(event_labels)  # Map event type numbers to labels

# Group data by month and item to count occurrences
monthly_counts = df.groupby(['month', 'item_label']).size().unstack(fill_value=0)

# Plotting the bar chart
plt.figure(figsize=(12, 8))

# Plot bars for each event type
months = monthly_counts.index
width = 0.25  # Bar width
x = range(len(months))

plt.bar(x, monthly_counts['call'], width=width, label='Call', color=sns.color_palette('Set2'))
plt.bar([i + width for i in x], monthly_counts['sms'], width=width, label='SMS')
plt.bar([i + 2 * width for i in x], monthly_counts['data'], width=width, label='Data')

# Adding labels and title
plt.title('Monthly Event Counts (Call, SMS, and Data)', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks([i + width for i in x], months, fontsize=12)  # Set month names as x-ticks
plt.legend(title='Event Type', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines for better readability
plt.tight_layout()

# Show the plot
plt.show()




````

#### Chart - 9 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
# Filter numeric columns
numeric_df = df.select_dtypes(include=["number"])

# Drop the 'item_category' column if it exists
if 'item_category' in numeric_df.columns:
    numeric_df = numeric_df.drop(columns=['item_category'])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Set the figure size
plt.figure(figsize=(10, 8))

# Create the heatmap
sns.heatmap(
    correlation_matrix,
    annot=True,           # Display correlation values on the heatmap
    fmt=".2f",            # Format for the correlation values
    cmap="coolwarm",      # Colormap for better visualization
    linewidths=0.5,       # Add lines between cells for clarity
    square=True           # Keep cells square-shaped
)

# Add a title to the heatmap
plt.title('Correlation Heatmap of Numeric Data', fontsize=16, fontweight='bold')

# Show the plot
plt.show()



````

##### 1. Why did you pick the specific chart?

A correlation heatmap is ideal for visualizing relationships between numerical variables. It provides a quick, intuitive way to identify:

**Strength and Direction of Relationships:** Shows positive/negative correlations.

**Patterns:** Highlights key variable interactions.

**Simplicity:** Compactly displays all pairwise correlations in one chart.

##### 2. What is/are the insight(s) found from the chart?

**Strong Relationships:** Identify variables with high positive or negative correlations, indicating strong dependencies.
