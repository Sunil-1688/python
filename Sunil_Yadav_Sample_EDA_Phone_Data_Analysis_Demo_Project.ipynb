## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Convert date column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Extract additional features from the date column
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour

# 1. Count of different `item` types
item_counts = df['item'].value_counts()
print("Counts of different item types:\n", item_counts)

# 2. Total duration of calls/messages/data usage by month
monthly_usage = df.groupby(['month', 'item'])['duration'].sum()
print("Total usage by month:\n", monthly_usage)

# 3. Average call/data duration per network
avg_duration_network = df.groupby('network')['duration'].mean()
print("Average duration per network:\n", avg_duration_network)

# 4. Most frequently used network
most_used_network = df['network'].value_counts().idxmax()
print("Most frequently used network:", most_used_network)

# 5. Peak usage hour
peak_hour = df['hour'].value_counts().idxmax()
print("Peak usage hour:", peak_hour)

# 6. Network preference for calls vs data
network_item_usage = df.groupby(['network', 'item'])['duration'].sum()
print("Network preference for calls and data:\n", network_item_usage)

# 7. Day with highest usage
day_highest_usage = df.groupby('day')['duration'].sum().idxmax()
print("Day with highest usage:", day_highest_usage)

# 8. Distribution of call durations
call_durations = df[df['item'] == 'call']['duration']
print("Summary of call durations:\n", call_durations.describe())

# 9. Identify periods with zero usage
zero_usage = df[df['duration'] == 0]
print("Periods with zero usage:\n", zero_usage)

# 10. Monthly trend of data usage
data_usage_trend = df[df['item'] == 'data'].groupby('month')['duration'].sum()
print("Monthly data usage trend:\n", data_usage_trend)
df.head()
````

#### Chart - 2 - Pie Cart

````python
# Chart - 2 visualization code
# Sample data preparation (replace this with your actual 'df' data)
item_counts = df['item'].value_counts()  # Count occurrences of each item type
labels = ['call', 'sms', 'data']  # Labels for the pie chart
counts = [item_counts[0], item_counts[1], item_counts[2]]  # Frequency counts for each category

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    counts,
    labels=labels,
    autopct='%1.1f%%',  # Show percentages
    colors=['green', 'red', 'orange'],  # Custom colors
    startangle=90,  # Rotate the start angle
    wedgeprops={'edgecolor': 'black'}  # Add black edges to wedges
)

# Add a title
plt.title('Proportion of Event Types (Call, SMS, Data)', fontsize=16)

# Display the pie chart
plt.show()

````

#### Chart - 12 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
# Filter numeric columns
numeric_df = df.select_dtypes(include=["number"])

# Drop the 'item_category' column if it exists
if 'item_category' in numeric_df.columns:
    numeric_df = numeric_df.drop(columns=['item_category'])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Set the figure size
plt.figure(figsize=(10, 8))

# Create the heatmap
sns.heatmap(
    correlation_matrix,
    annot=True,           # Display correlation values on the heatmap
    fmt=".2f",            # Format for the correlation values
    cmap="coolwarm",      # Colormap for better visualization
    linewidths=0.5,       # Add lines between cells for clarity
    square=True           # Keep cells square-shaped
)

# Add a title to the heatmap
plt.title('Correlation Heatmap of Numeric Data', fontsize=16, fontweight='bold')

# Show the plot
plt.show()



````

##### 1. Why did you pick the specific chart?

A correlation heatmap is ideal for visualizing relationships between numerical variables. It provides a quick, intuitive way to identify:

**Strength and Direction of Relationships:** Shows positive/negative correlations.

**Patterns:** Highlights key variable interactions.

**Simplicity:** Compactly displays all pairwise correlations in one chart.

##### 2. What is/are the insight(s) found from the chart?

**Strong Relationships:** Identify variables with high positive or negative correlations, indicating strong dependencies.

**Weak/No Relationships:** Spot variables with near-zero correlation, suggesting independence.

**Multicollinearity:** Detect highly correlated variables, which may need handling in regression models.

**Feature Selection:** Highlight variables most relevant to the target for predictive modeling.
